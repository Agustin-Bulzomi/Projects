```python
#se debe instalar ipyleaflet, mpu, openpyxl

import ipyleaflet
import mpu
import pandas as pd
import re
import numpy as np
from ipyleaflet import Map, basemaps, Marker, AwesomeIcon, Icon
```

```python
ruta_archivo = "properati.csv"
```

```python
df = pd.read_csv(ruta_archivo)
df
```

```python
campo_separado = df["place_with_parent_names"].str.split('|', expand = True)
campo_separado
```

```python
campo_separado.columns = ['??', 'Pais', 'Zona', 'Partido', 'Barrios', 'Country', 'Otra']
df = pd.concat([df, campo_separado], axis=1)
df
```

```python
df_limpio = df.drop(['description', 'geonames_id', 'lat-lon', 'floor', 'rooms', 'expenses', 'properati_url', 'image_thumbnail', 'title', '??'], axis = 1)
df_limpio.head()
```

```python
df_limpio['TC'] = df_limpio['price_aprox_local_currency'] / df_limpio['price_aprox_usd']
```

```python
df_limpio.head(10)
```

```python
df_limpio['TC'].mean().round(2)
```

```python
mascara_pilar = df_limpio['Partido'] == 'Pilar'
solo_pilar = df_limpio.loc[mascara_pilar, :]
solo_pilar
```

```python
# mascara para eliminar valores menores a 1
mascara_pilar_sin_precio = solo_pilar['price_aprox_usd'] > 0 # or np.isNan
ds_psp = solo_pilar.loc[mascara_pilar_sin_precio, :]
# usamos estas lineas para verificar si teniamos valores Nan
#ds_psp.sort_values(by = 'price_aprox_usd', ascending = False)
ds_psp.sort_values(by = 'price_aprox_usd', ascending = True)
```

```python
# eliminamos los nulos de la columna LAT y LON
ds_psp1 = ds_psp.dropna(subset=['lat', 'lon'])
ds_psp1.shape
```

```python
puentes_pilar = pd.DataFrame({
    'Nombre': ['Separacion', 'Tortugas', '26', 'Camaño', 'Palmas', '25', 'Manzanares','6'],
    'Latitud': [-34.468040, -34.448354, -34.440682, -34.436667, -34.443766, -34.449550, -34.441083, -34.395740],
    'Longitud': [-58.685014, -58.748577, -58.782740, -58.828145, -58.872848, -58.905720, -58.966898, -59.022886]
})
```

```python
print(puentes_pilar)
```

```python
mascara_barrios_vacios = ds_psp1['Barrios'] == ''
ds_psp2 = ds_psp1.loc[mascara_barrios_vacios, :]
ds_psp2.shape
```

```python
def distancia(lat, long):
    aux = []
    for index, row in puentes_pilar.iterrows():
        aux.append(
            mpu.haversine_distance(
                (row['Latitud'], row['Longitud']), 
                (lat, long)
            )
        )
    return min(aux)
ds_psp2['distancia'] = ds_psp2.apply(lambda x: distancia(x['lat'], x['lon']), axis=1)
ds_psp2
```

```python
ds_psp3 = ds_psp2[ds_psp2['price_usd_per_m2'].notna()]
ds_psp3
```

```python
ds_psp4 = ds_psp3[ds_psp3.index < 1500]
ds_psp4
```

```python
ds_psp4['quantile'] = pd.qcut(ds_psp4['price_usd_per_m2'], 4, labels=False)
ds_psp4
```

```python
def marca_propiedad(lat, long, quantile):
    if quantile == 0:
        icono_prop = icon0
    elif quantile == 1:
        icono_prop = icon1
    elif quantile == 2:
        icono_prop = icon2  
    else: 
        icono_prop = icon3
    marker = Marker(location=(lat,long), draggable=False, icon=icono_prop)
    basic_map.add_layer(marker)
```

```python
basic_map = Map(basemap=basemaps.OpenStreetMap.Mapnik, center=(-34.60823238941058, -58.398160858739075), zoom=14)

icono_subte = Icon(icon_url='icono_subte.png', icon_size=[28, 28])

icon0 = AwesomeIcon(
    name='home',
    marker_color='green',
    icon_color='black',
    spin=False
)

icon1 = AwesomeIcon(
    name='home',
    marker_color='beige',
    icon_color='black',
    spin=False
)

icon2 = AwesomeIcon(
    name='home',
    marker_color='orange',
    icon_color='black',
    spin=False
)

icon3 = AwesomeIcon(
    name='home',
    marker_color='red',
    icon_color='black',
    spin=False
)

for index, row in puentes_pilar.iterrows():
    marker = Marker(location=(row['Latitud'],row['Longitud']), draggable=False, icon=icono_subte)
    basic_map.add_layer(marker);
    
# comb4.apply(marca_propiedad(), axis=1)
ds_psp4[['lat','lon','quantile']].apply(lambda ds_psp4: marca_propiedad(ds_psp4['lat'],ds_psp4['lon'],ds_psp4['quantile']),axis=1)
    

display(basic_map)
```
