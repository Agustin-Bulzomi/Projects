```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

#Imports para Naive Bayes
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
votos = pd.read_csv('house-votes-84.csv')
votos.head(10)
```

```python
votos["Class Name"].value_counts(normalize = True).round(2)
```

```python
votos_numeric = votos.replace(to_replace = ["n", "?", "y", "democrat", "republican"], value = [-1, 0, 1, 0, 1])
votos_numeric.head(10)
```

```python
# Para medir la correlacion entre las distintas features y con la clase tambien

fig, ax = plt.subplots(figsize=(12,12))         
sns.heatmap(votos_numeric.corr(), ax=ax)
```

```python
X = votos.iloc[:,1:17]
y = votos["Class Name"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.preprocessing import OneHotEncoder 
```

```python
columnas_categoricas = [x for x in votos.columns if x != "Class Name"]
```

```python
encoder_categories = []
for col in columnas_categoricas:    
    col_categories = votos[col].unique()
    encoder_categories.append(col_categories)
```

```python
encoder = OneHotEncoder(categories = encoder_categories, sparse=False)
```

```python
encoder = encoder.fit(X_train[columnas_categoricas])
```

```python
print(encoder.get_feature_names())
```

```python
X_train_encoded = encoder.transform(X_train[columnas_categoricas])
X_train_categorical = pd.DataFrame(X_train_encoded, columns = encoder.get_feature_names(columnas_categoricas))


X_test_encoded = encoder.transform(X_test[columnas_categoricas])
X_test_categorical = pd.DataFrame(X_test_encoded, columns = encoder.get_feature_names(columnas_categoricas))
```

## Aca Termina el Preprocesamiento de los Datos y Comienza la aplicacion de los Modelos

```python
# Se crea la variable que contiene la regresión logística
logistic = LogisticRegression(C=1, penalty = "l2", solver = 'lbfgs')

# Se fitean los datos en el modelo
logistic.fit(X_train_categorical, y_train)
```

```python
# Se muestran los coeficientes de las features
logistic.coef_
```

```python
# Se predicen los valores
logistic_predictions = logistic.predict(X_test_categorical)
```

```python
logistic_predictions[:5,]
```

```python
logistic_predictions_proba = logistic.predict_proba(X_test_categorical)
```

```python
logistic_predictions_proba[:5,]
```

```python
# Se obtiene la accuracy
from sklearn.metrics import accuracy_score
acc_logistic = accuracy_score(y_test, logistic_predictions)

print(f"accuracy_score: {acc_logistic.round(2)}")
```

```python
grid_search_logistic = GridSearchCV(LogisticRegression(), {'C': [0.1, 1, 10]}, cv=3, scoring='accuracy')
grid_search_logistic.fit(X_train_categorical, y_train)

display(grid_search_logistic.best_estimator_,grid_search_logistic.best_score_)
```

```python
cm_votos = confusion_matrix(y_test, logistic_predictions)
plt.plot(30)
ax = sns.heatmap(cm_votos, annot=True, cmap='Blues', cbar=False,
                    annot_kws={"size": 14}, fmt='g')

plt.title('Matriz de confusión', size =16)
plt.xlabel('Valores previstos', size = 13)
plt.ylabel('Valores verdaderos', size=13)
```
