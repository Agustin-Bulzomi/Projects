```python
def limpieza_amb(ambientes_desc, ambientes_t):
    if pd.isnull(ambientes_desc) and pd.isnull(ambientes_t):
        ambientes = np.NaN
    elif pd.isnull(ambientes_desc):
        ambientes = ambientes_t
    else:
        ambientes = ambientes_desc
    return ambientes
```

```python
data_concat["ambientes"] = data_concat.apply(lambda data_concat: limpieza_amb(data_concat['ambientes_desc'],data_concat['ambientes_t']),axis=1)
#data_concat.head(15)
```

```python
def limpieza_amb2(rooms, ambientes):
    if pd.isnull(rooms) and pd.isnull(ambientes):
        ambientes_train = 0
    elif pd.isnull(rooms):
        ambientes_train = ambientes
    else:
        ambientes_train = int(rooms)
    return int(ambientes_train)
```

```python
data_concat["ambientes_train"] = data_concat.apply(lambda x: limpieza_amb2(x['rooms'],x['ambientes']),axis=1)
```

# Imputación

### Divisores de ambientes

```python
amb_1 = data_sin_columnas['ambientes_train'] == 1
amb_2 = data_sin_columnas['ambientes_train'] == 2
amb_3 = data_sin_columnas['ambientes_train'] == 3
amb_4 = data_sin_columnas['ambientes_train'] == 4
amb_5 = data_sin_columnas['ambientes_train'] == 5
amb_6 = data_sin_columnas['ambientes_train'] == 6
amb_7 = data_sin_columnas['ambientes_train'] == 7

divisor1 = (data_sin_columnas[amb_1].surface_covered_in_m2.median() + data_sin_columnas[amb_2].surface_covered_in_m2.median())/2
divisor2 = (data_sin_columnas[amb_2].surface_covered_in_m2.median() + data_sin_columnas[amb_3].surface_covered_in_m2.median())/2
divisor3 = (data_sin_columnas[amb_3].surface_covered_in_m2.median() + data_sin_columnas[amb_4].surface_covered_in_m2.median())/2
divisor4 = (data_sin_columnas[amb_4].surface_covered_in_m2.median() + data_sin_columnas[amb_5].surface_covered_in_m2.median())/2
divisor5 = (data_sin_columnas[amb_5].surface_covered_in_m2.median() + data_sin_columnas[amb_6].surface_covered_in_m2.median())/2
divisor6 = (data_sin_columnas[amb_6].surface_covered_in_m2.median() + data_sin_columnas[amb_7].surface_covered_in_m2.median())/2
divisor7 = (data_sin_columnas[amb_7].surface_covered_in_m2.median() + data_sin_columnas[amb_7].surface_covered_in_m2.max())/2
```

```python
# Función para asignar ambientes según superficie en registros cuyo valor para cantidad de ambientes es nulo
def asignar_ambientes_segun_superficie(surface_covered_in_m2):
    #if superficie.isnull():
    #    return 0
    #elif 0 < superficie <= divisor1:
    if 0 < surface_covered_in_m2 <= divisor1:
        return 1
    elif divisor1 < surface_covered_in_m2 <= divisor2:
        return 2
    elif divisor2 < surface_covered_in_m2 <= divisor3:
        return 3
    elif divisor3 < surface_covered_in_m2 <= divisor4:
        return 4
    elif divisor4 < surface_covered_in_m2 <= divisor5:
        return 5
    elif divisor5 < surface_covered_in_m2 <= divisor6:
        return 6
    elif divisor6 < surface_covered_in_m2 <= divisor7:
        return 7
    else:
        return np.NaN
```

```python
data_sin_columnas["ambientes_touched"] = data_sin_columnas.apply(lambda x: asignar_ambientes_segun_superficie(x['surface_covered_in_m2']) if int(x['ambientes_train']) == 0 else 0,axis=1)
```

```python
data_sin_columnas["ambientes_final"] = data_sin_columnas["ambientes_train"] + data_sin_columnas["ambientes_touched"]
```
