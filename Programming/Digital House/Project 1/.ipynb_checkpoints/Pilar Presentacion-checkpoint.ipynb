```python
import ipyleaflet
import mpu
import pandas as pd
import re
import numpy as np
from ipyleaflet import Map, basemaps, Marker, AwesomeIcon, Icon
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# leemos el archivo de Properati
location = "properati.csv"
data = pd.read_csv(location)
data.head()
```

**AGREGADO DE COLUMNAS**

```python
# separamos la columna "place_with_parent_names" por '|' para obtener info adicional
separado = data["place_with_parent_names"].str.split('|', expand = True)
separado.head()
```

```python
# agregamos dichas columnas del paso anterior al DF original
separado.columns = ['??', 'Pais', 'Zona', 'Partido', 'Barrios', 'Country', 'Otra']
data_concat = pd.concat([data, separado], axis=1)
#data_concat
```

```python
# calculamos el tipo de cambio para corroborar que en todos los registros sea el mismo
data_concat['TC'] = data_concat['price_aprox_local_currency'] / data_concat['price_aprox_usd'] 
# calculando el promedio, decidimos tomar la columna 'price_aprox_usd' como el $ de las propiedades
data_concat['TC'].mean().round(2)
```

```python
# funcion que nos permite limpiar la columna 'surface_covered_in_m2'
# se eligio dicha columna con respecto a 'surface_total_in_m2' ya que tenia menor cantidad de registros nulos
def limpieza_superficie(sup_total, sup_cubierta):
    if sup_total is not None and sup_cubierta is not None:
            if sup_total >= sup_cubierta:
                superficie = sup_total
            else: 
                superficie = np.NaN
    elif sup_total is not None:
            superficie = sup_total
    elif sup_cubierta is not None:
            superficie = sup_cubierta
    else: superficie = np.NaN        
                    
    return superficie
```

```python
# creacion de una columna nueva de m2 utilizando la funcion 'limpieza_superficie'
data_concat["superficie"] = data_concat[['surface_total_in_m2','surface_covered_in_m2']].apply(lambda data_concat: limpieza_superficie(data_concat['surface_total_in_m2'],data_concat['surface_covered_in_m2']),axis=1)

data_concat.head(10)
```

```python
# creacion de la columna definitiva para obtener el $/m2
data_concat['Precio_USD_por_M2']=data_concat.price_aprox_usd/data_concat.superficie 
data_concat.head()
```

**ELIMINADO DE COLUMNAS**

```python
# eliminamos las columnas que no corresponden a nuestro analisis (algunas por no necesitarlas, otras por estar repetidas)
data_eliminacion = data_concat.drop(['Unnamed: 0', 'operation', 'place_with_parent_names', 'place_name', 'country_name', 'state_name',
                                     'description', 'geonames_id', 'lat-lon', 'floor', 'rooms', 'expenses', 'properati_url', 
                                     'image_thumbnail', 'title', '??'], axis = 1)
data_eliminacion.head()
```

```python
# nuevo DF limpio
data_eliminacion.shape
```

```python
# mascara para obtener dentro de Pilar, los NO country's
pilar_partido = solo_pilar.isin({'Barrios':['Pilar', 'Del Viso', 'Derqui', 'Villa Rosa', 'Villa Stolfi', 'Manuel Alberti', 'Manzanares'] })
pilar_sincountry = solo_pilar[pilar_partido['Barrios']]
pilar_sincountry.head()
```
